```python
# After thoroughly analyzing the data, this column seems to provide the most accurate data on 
# whether or not a company received vc funding

df1['Last Equity Funding Type'].value_counts()
```

```python
# Create an empty column for the label
df1['Label']= np.nan
```

```python
# To set up our label for a logistic regression model, make the label of every company that got VC funding a value of 1

df1['Label'][(df['Last Equity Funding Type'] == 'Venture - Series Unknown') | 
               (df['Last Equity Funding Type'] == 'Series A') |
               (df['Last Equity Funding Type'] == 'Series B') |
               (df['Last Equity Funding Type'] == 'Series C') |
               (df['Last Equity Funding Type'] == 'Series D') |
               (df['Last Equity Funding Type'] == 'Post-IPO Equity') |
               (df['Last Equity Funding Type'] == 'Series E') |
               (df['Last Equity Funding Type'] == 'Series F')] = 1
    
```

```python
# Now to set a value of 0 to every company that did not get venture funding
df1['Label'].fillna(0, inplace=True)
```

```python
#Now reinsert the Nans
df1['Label'][df1['Last Equity Funding Type'].isna() == True] = np.nan
```

```python
#This gives us a total dataset of 70108 rows
df1['Label'].value_counts().sum()
```

```python
# Now to clear the rows that have missing label values
dropped_na = df1.dropna(subset=['Label'])
dropped_na.shape
```
